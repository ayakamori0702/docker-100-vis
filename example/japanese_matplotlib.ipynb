### **豆腐になっているのを直す**

font_managerから適切なフォントを探す

```python
from matplotlib import font_manager
for i in font_manager.fontManager.ttflist:
    if ".ttc" in i.fname:  # フォントの詰め合わせファイルのようなもの
        print(i)
```

```python
plt.rc('font', family='Songti SC')
plt.figure(figsize=(10, 15))
x = tmp.index
y = tmp.values
plt.barh(x, y)
```
